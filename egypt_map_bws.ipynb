```python
import geopandas as gpd
```

```python
file = gpd.read_file('data_egy.gpkg')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
file.plot(column = 'bws_score', ax=ax, legend=True, scheme='quantiles')
#ax.legend(title="Baseline Water Stress")
ax.set_title("Baseline Water Stress (score 1-5) - Egypt", fontsize=25)
ax.set_axis_off()
plt.savefig('egypt_map_bws.png')
```
